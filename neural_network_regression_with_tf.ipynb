```python
#Evaluating a model
X=tf.range(-100,100,4)
```

```python
y=X+10
X,y
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# train test validation set
X_train =X[:40]
y_train=y[:40]
X_test=X[40:]
y_test=y[40:]
len(X_train)
```

```python
len(X_test)
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X_train,y_train,c="b")
plt.scatter(X_test,y_test,c="g")
plt.legend()
```

```python
model=Sequential([tf.keras.layers.Dense(1000,input_dim=1,activation="relu"),
                  tf.keras.layers.Dense(764,input_dim=1,activation="relu"),
tf.keras.layers.Dense(1,input_dim=1,activation="linear")])


model.compile(loss="mse",
              optimizer="Adam",
              metrics=["mae"]
              )
model.fit(X_train,y_train,epochs=100)
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model=model,show_shapes=True)
```

```python
y_pred=model.predict(X_test)
```

```python
plt.plot(y_test,c="g")
plt.plot(y_pred,c="b")
```
